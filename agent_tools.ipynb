```python
from langchain.chains import LLMMathChain
from langchain.agents import Tool
```

```python
llm_math=LLMMathChain(llm=llm)
math_tool =  Tool(
    name='calculator',
    func=llm_math.run,
    description='Useful when you need to answer questions about math'
)

tools=[math_tool]
```

# Creating Agent

### Zero Shot

```python
from langchain.agents import initialize_agent

zero_shot_agent = initialize_agent(
    agent="zero-shot-react-description",
    tools=tools,
    llm=llm,
    Verbose=True,
    max_iteration=10,
    handle_parsing_errors=True
)
```

```python
zero_shot_agent("What is root over 25")
```

```python
from langchain.tools import DuckDuckGoSearchRun

search = DuckDuckGoSearchRun()

search_tool = Tool.from_function(
    func=search.run,
    name="Search",
    description="useful when you need to search information from internet"
)

llm_math_chain = LLMMathChain(llm=llm,verbose=True)

math_tool =  Tool.from_function(
    func=llm_math_chain.run,
    name="Calculator",
    description="Useful when you are asked to perform mathematics"
)
```

```python
tools=[search_tool, math_tool]
agent=initialize_agent(
    tools,
    llm,
    agent="zero-shot-react-description",
    verbose=True
)
```

```python
agent.run("""Get Microsoft Stock price and display it in both USD and INR values""")
```

## Create a Custom tool

```python
from langchain_core.prompts import PromptTemplate
from langchain.tools import BaseTool
from langchain_core.output_parsers import JsonOutputParser
from langchain.tools import tool

@tool("lower_case", return_direct=True)
def StructuredResponseTool(question:str):
    """
    use this tool to send a prompt and get a json returned with three fields - Topic, Question_Details and Detailed_Response
    """
    json_prompt= PromptTemplate.from_template(
        """
        Return a JSON object with an `answer` key that answers the following question : {question}.
        The JSON object will have three fields - Topic, Question_Details and Detailed_Response
        """
    )

    model = GoogleGenerativeAI(model="gemini-1.5-pro-latest",temperature=0.5)

    json_parser = JsonOutputParser()
    json_chain = json_prompt | model | json_parser
    x=json_chain.invoke({"question":question})
    return x
```

```python
from langchain.chat_models import google_palm
from langchain.prompts import PromptTemplate
from langchain.chains.conversation.memory import ConversationBufferWindowMemory
```

```python
from langchain.agents import initialize_agent

tools = [StructuredResponseTool]
zero_shot_agent = initialize_agent(
    agent="zero-shot-react-description",
    tools=tools,
    llm=llm,
    verbose=True,
    max_iterations=10
)
```

```python
zero_shot_agent.run(input="Which is the top football team in world and how")
```
