## Getting Deeper into Agents

### Conversational ReAct Agents

```python
from langchain.agents import create_react_agent, load_tools
from langchain_google_genai import GoogleGenerativeAI
from langchain.memory import ConversationBufferMemory
from dotenv import load_dotenv
load_dotenv()
```

```python
llm = GoogleGenerativeAI(model="gemini-1.5-pro-latest",temperature=0.5)
```

```python
tools = load_tools(["llm-math"], llm=llm)
```

```python
from langchain.agents import initialize_agent
memory = ConversationBufferMemory(memory_key="chat_history")

conversational_agent = initialize_agent(
    agent="conversational-react-description",
    tools=tools,
    llm=llm,
    verbose=True,
    max_iterations=3,
    memory=memory
)
```

```python
output_1= conversational_agent.run("Add 7 to 9 and tell me the result")
```

```python
output_2=conversational_agent.run("Add 5 to previous result")
```

### ReAct DocStore

```python
from langchain.agents import initialize_agent, Tool
from langchain import Wikipedia
from langchain.agents.react.base import DocstoreExplorer
```

```python
%pip install wikipedia
```

```python
docstore = DocstoreExplorer(Wikipedia())
tools=[
    Tool(name="Search", func=docstore.search, description="Useful when you need to search from wikipedia"),
    Tool(name="Lookup", func=docstore.lookup, description="Useful when you need to lookup a term in wikipedia")
]

llm= GoogleGenerativeAI(model="gemini-1.5-pro-latest", temperature=0)

docstore_agent=initialize_agent(
    tools,
    llm,
    agent="react-docstore",
    verbose=True,
    handle_parsing_errors=True,
    max_iterations=8
)
```

```python
docstore_agent("Tell me few key things about Bhagat Singh")
```

## Self Ask with Search

```python
from langchain.tools import DuckDuckGoSearchResults
search = DuckDuckGoSearchResults()

tools=[
    Tool(
        func=search.run,
        name="Intermediate Answer",
        description="Useful when you need to search the internet for information"
    )
]

agent=initialize_agent(
    tools=tools,
    llm=llm,
    agent="self-ask-with-search",
    verbose=True,
    handle_parsing_errors=True
)
```

```python
print(agent.invoke("Question : Who was the president of USA when the first moon landing took place?"))
```
